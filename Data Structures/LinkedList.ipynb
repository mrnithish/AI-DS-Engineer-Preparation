```python
# Node Creation
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

# Linked List 
class LinkedList:

    #Head initializing
    def __init__(self):
        self.head=None

    #Insert Begin
    def insertAtBegin(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            new_node.next=self.head
            self.head=new_node

    #Insert At End
    def insertAtEnd(self,data):
        new_node=Node(data)
        node=self.head
        if self.head  is  None:
            self.head=new_node
        else:
            while node.next is not None: # It needs to be stopped at the last So use node.next
                node=node.next
            node.next=new_node

    #Insert the element after the given value
    def insert_after(self,data,x):
        node=self.head
        while node  is not None:
            if x==node.data:
                break
            else:
                node=node.next
        if node  is None:
            print("The Given Value is not Present in the LinkedList")
        else:
            new_node=Node(data)
            new_node.next=node.next
            node.next=new_node
    

    #Insert the element before the given value
    def insert_before(self,data,x):
        new_node=Node(data)
        node=self.head
        if node is None:
            print("List is Empty")
        if node.data == x:
            new_node.next=node
            node=new_node
        else:
            while node.next is not None:
                if node.next.data==x:
                    break
                node=node.next
            if node.next is None:
                print("Not Exist")
            else:
                new_node.next=node.next #Similar insert after element logic
                node.next=new_node
    
    #Insert at the Postion
    def insert_position(self,data,pos):
        new_node=Node(data)
        node=self.head
        if node is None:
            print("List is Empty")
        if pos==0:
            new_node.next=node
            node=new_node
        else:
            count=1
            while node is not None:
                if count==pos: # It places the specific postion of index
                    break
                count+=1
                node=node.next
            if node is None:
                print("Index Not Found")
            else:
                new_node.next=node.next
                node.next=new_node

    #Delete at Begin
    def deleteAtBegin(self):
        if self.head is None:
            print("List is Empty")
        else:
            self.head=self.head.next
    
    #Delete at the End
    def deleteAtEnd(self):
        if self.head is None:
            print("List is Empty")
        elif self.head.next is None:
            self.head=None
        else:
            node=self.head
            while node.next.next is not None:
                node=node.next
            node.next=None

    #delete by value
    def deleteByValue(self,value):
        node=self.head
        if node is None:
            print("List is Empty")
        elif node.data is value:
            self.head=self.head.next
        else:
            while node.next is not None:
                if node.next.data== value:
                    break
                node=node.next
            if node.next is None:
                print("Not present in the List")
            else:
                node.next=node.next.next

    #Searching
    def searchElement(self,value):
        node=self.head
        if self.head is None:
            print("List is Empty")
        elif node.data == value:
            print("Element Found at 0th position")
        else:
            pos=1
            while node.next != None:
                if node.data==value:
                    break
                pos+=1
                node=node.next
            if node.next is None:
                print("Element Not Present")
            else:
                print(f'{value} is found at {pos}th position')

    #Get the Middle  Element in a list
    def getMiddle(self):
        node=self.head
        slow,fast=node,node.next
        while fast is not None and fast.next is not None:
            slow=slow.next
            fast=fast.next.next
        print(slow.data)
    




    # Traversal
    def display(self):
        node=self.head
        if node is None:
            print("List is Empty")
        else:
            while node is not None: # We need to traverse via the last node
                print(node.data,end=" -> ")
                node=node.next
            print()

    def reverse(self):
        curr=self.head
        prev=None
        if curr is None:
            print("List is Empty")
        else:
            while curr is not None:
                temp=curr.next
                curr.next=prev
                prev=curr
                curr=temp
        self.head=prev
        return prev
                



obj=LinkedList()
obj.insertAtEnd(0)
obj.insertAtBegin(10)
obj.insertAtBegin(20)
obj.insertAtEnd(34)
obj.insert_after(400,34)
obj.insert_before(4,10)
obj.insert_position(100,2)
obj.display()
obj.deleteAtBegin()
obj.display()
obj.deleteAtEnd()
obj.display()
obj.searchElement(10)
obj.getMiddle()
obj.display()
obj.reverse()
obj.display()
```

Given head, the head of a linked list, determine if the linked list has a cycle in it.

There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to. Note that pos is not passed as a parameter.

Return true if there is a cycle in the linked list. Otherwise, return false.

 

Example 1:


Input: head = [3,2,0,-4], pos = 1
Output: true
Explanation: There is a cycle in the linked list, where the tail connects to the 1st node (0-indexed).
Example 2:


Input: head = [1,2], pos = 0
Output: true
Explanation: There is a cycle in the linked list, where the tail connects to the 0th node.
Example 3:


Input: head = [1], pos = -1
Output: false
Explanation: There is no cycle in the linked list.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        curr=head
        visited=set()
        while curr:
            if curr in visited:
                return True
            visited.add(curr)
            curr=curr.next
        return False

```
